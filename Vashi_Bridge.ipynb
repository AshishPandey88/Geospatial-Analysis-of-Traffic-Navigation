# **Import the required packages**

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.draw import disk
from skimage.morphology import (erosion, dilation, closing, opening,
                                area_closing, area_opening)
from skimage.color import rgb2gray
from math import sqrt
```

```python
vashi = imread('Vashi_hd.jpg')
```

**#Convert the image to Binary**

```python
fig, ax = plt.subplots(1,2, figsize=(20,20))
ax[0].imshow(vashi);
ax[0].set_title('Original')

binary = rgb2gray(vashi)
ax[1].imshow(binary)
ax[1].set_title('Binary')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.feature import blob_dog, blob_log, blob_doh
```

Laplacian of Gaussian (LoG) on this binarized image
We will be only setting the min_sigma parameter on all methods. 
The rest will be the default parameter setting.

#Laplacian of Gaussian (LoG)

```python
vashi_traffic = blob_log(binary, min_sigma=0.9
                         ,max_sigma=30
                         ,threshold=0.15
                         ,overlap=0.05
                         ,exclude_border=50
                         )
```

```python
len(vashi_traffic)
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.imshow(vashi)
for blob in vashi_traffic:
    y, x, area = blob
    ax.add_patch(plt.Circle((x, y), area*np.sqrt(2), 
                 color='r', fill=False))
    ax.set_title('Satellite Image analysis- Traffic count on Mumbai Vashi Toll')
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax.imshow(vashi)
for blob in vashi_traffic:
    y, x, area = blob
    if x<1150:
      ax.add_patch(plt.Circle((x, y), area*np.sqrt(2), 
                 color='yellow', fill=False))
      plt.text(1000, 380, "Vashi Bridge Toll", bbox=dict(fill=True,facecolor='yellow', edgecolor='white', linewidth=2))
      plt.text(1400, 1000,('No. of Cars- Crossed the Toll',len(vashi_traffic[vashi_traffic[:,1]<1150])), bbox=dict(fill=True,facecolor='yellow', edgecolor='white', linewidth=2))
      #plt.text(700, 500, len(vashi_traffic[vashi_traffic[:,1]<472]), bbox=dict(fill=True,facecolor='yellow', edgecolor='white', linewidth=2))
    else:
      ax.add_patch(plt.Circle((x, y), area*np.sqrt(2), 
                 color='r', fill=False))
      plt.text(1350, 1050, ('No. of Cars-heading towards the Toll',len(vashi_traffic[vashi_traffic[:,1]>1150])),color='red', bbox=dict(fill=True,facecolor='white', edgecolor='white', linewidth=2))
    ax.set_title('Satellite Image analysis- Traffic count on Mumbai Vashi Toll')
```
